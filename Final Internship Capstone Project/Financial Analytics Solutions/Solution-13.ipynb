```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')
```

```python
# Add a new column called â€˜Rankâ€™ that assigns a rank to each company based on their revenue (highest to lowest)
data['Rank'] = data['Revenue'].rank(ascending=False)
```

```python
# Identify the company with the highest rank
company_highest_rank = data.loc[data['Rank'].idxmin(), 'Company']
highest_rank_value = data['Rank'].min()

print(f"The company with the highest rank is {company_highest_rank} with a rank of {highest_rank_value}")
```

```python
import plotly.graph_objects as go

# Sort data by rank
data_sorted = data.sort_values('Rank')

# Create a bar plot with Plotly
fig = go.Figure()

# Add bar trace
fig.add_trace(go.Bar(
    x=data_sorted['Company'],
    y=data_sorted['Revenue'],
    hovertext=data_sorted['Rank'],
    hoverinfo='text',
    marker=dict(color=data_sorted['Revenue'], colorscale='Viridis', colorbar=dict(title='Revenue')),
    name='Revenue'
))

# Add layout settings
fig.update_layout(
    title='Revenue Rankings',
    xaxis=dict(title='Company', tickangle=45),
    yaxis=dict(title='Revenue'),
    barmode='group',
    showlegend=False,
    template='plotly_dark'  # Dark theme
)

# Add hover information
fig.update_traces(hoverinfo='text', hovertext='<b>Company:</b> ' + data_sorted['Company'] +
                  '<br><b>Revenue:</b> $' + data_sorted['Revenue'].astype(str) +
                  '<br><b>Rank:</b> ' + data_sorted['Rank'].astype(str))

# Show the interactive plot
fig.show()
```

**Infer findings:**

1. **Company with Highest Rank:**
   - The company 'A' has the highest rank of 1.0 based on revenue.

2. **Revenue Rankings of Companies:**
   - The bar plot shows the revenue rankings of companies, with higher bars indicating higher revenue.
   - Analyzing this can provide insights into the revenue performance of each company.
