```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')
```

```python
# Identify outliers in the dividend payments
dividend_outliers = data[(data['Dividends'] < data['Dividends'].quantile(0.01)) | (data['Dividends'] > data['Dividends'].quantile(0.99))]
print(f"Number of outliers in dividend payments: {len(dividend_outliers)}")
print("Outliers in dividend payments:")
print(dividend_outliers[['Company', 'Dividends']])
```

```python
# Visualize the distribution of dividend payments
fig = px.box(data, y='Dividends', points='outliers', title='Distribution of Dividend Payments')
fig.update_layout(yaxis_title='Dividend Payments')
fig.show()
```

```python
# Analyze the impact of outliers on the distribution
without_outliers = data[(data['Dividends'] >= data['Dividends'].quantile(0.01)) & (data['Dividends'] <= data['Dividends'].quantile(0.99))]
fig = px.histogram(without_outliers, x='Dividends', nbins=30, title='Distribution of Dividend Payments (Without Outliers)')
fig.update_layout(xaxis_title='Dividend Payments', yaxis_title='Count')
fig.show()
```

```python
# Investigate potential reasons for outliers
print("Investigating potential reasons for outliers...")
for outlier in dividend_outliers.itertuples():
    company = outlier.Company
    dividend = outlier.Dividends
    print(f"Company: {company}, Dividend Payment: {dividend}")
    company_data = data[data['Company'] == company]
    print(f"Revenue: {company_data['Revenue'].values[0]}, Profit: {company_data['Profit'].values[0]}")
    print("Potential reasons could be exceptional performance, special dividends, or data errors.")
    print("-" * 50)
```


Infer findings:
1. Outliers in Dividend Payments:
   There are no outliers in the dividend payments column.
2. Impact of Outliers on Distribution:
   Outliers can skew the overall distribution of dividend payments, leading to a distorted view of the typical dividend payment range.
   Understanding and handling outliers is important for accurate analysis and decision-making.
