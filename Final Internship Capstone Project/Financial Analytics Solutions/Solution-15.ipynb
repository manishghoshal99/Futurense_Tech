```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')
```

```python
# Create a box plot for profit distribution across industries
fig = px.box(data, x='Industry', y='Profit', title='Profit Distribution Across Industries',
             color='Industry', template='plotly_dark')
```

```python
# Add layout settings
fig.update_layout(xaxis_title='Industry', yaxis_title='Profit', showlegend=False)

# Show the interactive plot
fig.show()
```
