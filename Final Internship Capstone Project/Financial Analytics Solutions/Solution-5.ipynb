```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')

# Calculate the debt-to-equity ratio
data['DebtToEquityRatio'] = data['Liabilities'] / data['Equity']
```

```python
# Find the company with the highest debt-to-equity ratio
highest_debt_to_equity = data['DebtToEquityRatio'].max()
company_with_highest_debt_to_equity = data.loc[data['DebtToEquityRatio'] == highest_debt_to_equity, 'Company'].values[0]
print(f"The company with the highest debt-to-equity ratio is {company_with_highest_debt_to_equity} with a ratio of {highest_debt_to_equity:.2f}")
```

```python
# Calculate summary statistics for debt-to-equity ratio by industry
debt_to_equity_summary = data.groupby('Industry')['DebtToEquityRatio'].describe()
print(debt_to_equity_summary)
```
