```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')

# Calculate the debt-to-equity ratio
df['DebtToEquityRatio'] = df['Liabilities'] / df['Equity']
```

```python
# Calculate the net income for each company
df['NetIncome'] = df['Revenue'] - df['Expenses']
```

```python
# Find the company with the highest net income
highest_net_income_company = df.loc[df['NetIncome'].idxmax(), 'Company']
highest_net_income = df['NetIncome'].max()
print("Company with Highest Net Income:", highest_net_income_company)
print("Highest Net Income:", highest_net_income)
```

```python
# Calculate the correlation between net income and market capitalization
correlation_net_income_marketcap = df['NetIncome'].corr(df['MarketCap'])
print("Correlation between Net Income and Market Capitalization:", correlation_net_income_marketcap)
```

```python
# Visualize the relationship between net income and market capitalization using a scatter plot
fig = px.scatter(df, x='NetIncome', y='MarketCap', title='Relationship between Net Income and Market Capitalization',
                 labels={'NetIncome': 'Net Income', 'MarketCap': 'Market Capitalization'})
fig.show()
```
