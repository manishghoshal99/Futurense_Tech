```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')

# Calculate the debt-to-equity ratio
data['DebtToEquityRatio'] = data['Liabilities'] / data['Equity']
```

```python
# Calculate the net income
data['NetIncome'] = data['Revenue'] - data['Expenses']
```

```python
# Find the company with the highest net income
highest_net_income_company = data.loc[data['NetIncome'] == data['NetIncome'].max(), 'Company'].values[0]
highest_net_income_value = data.loc[data['Company'] == highest_net_income_company, 'NetIncome'].values[0]
print(f"The company with the highest net income is {highest_net_income_company} with a net income of {highest_net_income_value}")
```

```python
# Visualize the relationship between net income and market capitalization
fig = px.scatter(data, x='NetIncome', y='MarketCap', color='Industry', hover_name='Company',
                 title='Net Income vs. Market Capitalization')
fig.update_layout(xaxis_title='Net Income', yaxis_title='Market Capitalization')
fig.show()
```

```python
# Calculate the correlation between net income and market capitalization
net_income_market_cap_corr = data['NetIncome'].corr(data['MarketCap'])
print(f"Correlation between net income and market capitalization: {net_income_market_cap_corr:.2f}")
```
