```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')
```

```python
# Calculate the Return on Investment (ROI)
data['ROI'] = data['Profit'] / data['Assets']
```

```python
# Find the company with the highest ROI
highest_roi_company = data.loc[data['ROI'] == data['ROI'].max(), 'Company'].values[0]
highest_roi_value = data.loc[data['Company'] == highest_roi_company, 'ROI'].values[0]
print(f"The company with the highest ROI is {highest_roi_company} with an ROI of {highest_roi_value:.2f}")
```

```python
# Visualize the relationship between ROI and market capitalization
fig = px.scatter(data, x='ROI', y='MarketCap', color='Industry', hover_name='Company',
                 title='ROI vs. Market Capitalization')
fig.update_layout(xaxis_title='Return on Investment (ROI)', yaxis_title='Market Capitalization')
fig.show()
```

```python
# Calculate the correlation between ROI and market capitalization
roi_market_cap_corr = data['ROI'].corr(data['MarketCap'])
print(f"Correlation between ROI and market capitalization: {roi_market_cap_corr:.2f}")
```
