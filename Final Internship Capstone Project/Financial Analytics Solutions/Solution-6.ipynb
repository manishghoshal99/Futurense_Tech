```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')
```

```python
# Calculate the Return on Investment (ROI)
data['ROI'] = data['Profit'] / data['Assets']
```

```python
# Find the company with the highest ROI
highest_roi_company = data.loc[data['ROI'] == data['ROI'].max(), 'Company'].values[0]
highest_roi_value = data.loc[data['Company'] == highest_roi_company, 'ROI'].values[0]
print(f"The company with the highest ROI is {highest_roi_company} with an ROI of {highest_roi_value:.2f}")
```

```python
# Create a scatter plot of ROI vs MarketCap
fig = px.scatter(data, y='MarketCap', x='ROI', color='Industry', size='Assets',
                 title='Return on Investment (ROI) vs Market Capitalization',
                 labels={'MarketCap': 'Market Capitalization', 'ROI': 'Return on Investment'},
                 hover_data=['Company', 'Industry', 'Profit', 'Assets', 'ROI'])

# Customize the layout
fig.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
    font_color='white',
    title_font_color='white',
    xaxis_title_font_color='white',
    yaxis_title_font_color='white',
    legend_title_font_color='white',
    legend_font_color='white'
)

# Identify the company with the highest ROI
highest_roi_company = data.loc[data['ROI'].idxmax(), 'Company']
highest_roi_value = data.loc[data['ROI'].idxmax(), 'ROI']

# Add annotation for the company with the highest ROI
fig.add_annotation(
    x=highest_roi_value,
    y=data.loc[data['ROI'].idxmax(), 'MarketCap'],
    text=f"* {highest_roi_company} had the highest ROI of {highest_roi_value:.2f}",
    showarrow=True,
    arrowhead=1,
    font=dict(color='yellow', size=14),
    bgcolor='rgba(0, 0, 0, 0.8)'
)

# Display the interactive plot
fig.show()
```

```python
# Calculate the correlation between ROI and market capitalization
roi_market_cap_corr = data['ROI'].corr(data['MarketCap'])
print(f"Correlation between ROI and market capitalization: {roi_market_cap_corr:.2f}")
```
