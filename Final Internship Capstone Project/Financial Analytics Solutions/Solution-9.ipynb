```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')
```

```python
# Calculate the correlation between revenue and expenses
overall_corr = data['Revenue'].corr(data['Expenses'])
print(f"Overall correlation between revenue and expenses: {overall_corr:.2f}")
```

```python
# Visualize the relationship between revenue and expenses
fig = px.scatter(data, x='Revenue', y='Expenses', color='Industry', hover_name='Company',
                 title='Revenue vs. Expenses')
fig.update_layout(xaxis_title='Revenue', yaxis_title='Expenses')
fig.show()
```

```python
# Calculate the correlation between revenue and expenses by industry
industry_corr = data.groupby('Industry').apply(lambda x: x['Revenue'].corr(x['Expenses'])).reset_index(name='Correlation')
print("Correlation between revenue and expenses by industry:")
print(industry_corr)
```

```python

# Visualize the relationship between revenue and expenses by industry
fig = px.scatter(data, x='Revenue', y='Expenses', color='Industry', facet_col='Industry',
                 hover_name='Company', title='Revenue vs. Expenses by Industry')
fig.update_layout(xaxis_title='Revenue', yaxis_title='Expenses')
fig.show()

```
