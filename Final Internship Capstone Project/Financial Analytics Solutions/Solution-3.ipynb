```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')
```

```python
import pandas as pd
import plotly.express as px


# Create a heatmap of EPS vs. P/E ratio
fig = px.density_heatmap(data, x='EarningsPerShare', y='PriceEarningsRatio', nbinsx=20, nbinsy=20,
                         color_continuous_scale='Viridis', title='Earnings per Share vs. Price-Earnings Ratio (Heatmap)')
fig.update_layout(xaxis_title='Earnings per Share (EPS)', yaxis_title='Price-Earnings Ratio (P/E)',
                  coloraxis_colorbar=dict(title='Density'))
fig.show()
```

```python
# Calculate the correlation between EPS and P/E ratio
eps_pe_corr = data['EarningsPerShare'].corr(data['PriceEarningsRatio'])
print(f"Correlation between Earnings per Share and Price-Earnings Ratio: {eps_pe_corr:.2f}")

# Interpret the correlation
if eps_pe_corr > 0.5:
    print("There is a strong positive correlation between EPS and P/E ratio.")
elif eps_pe_corr < -0.5:
    print("There is a strong negative correlation between EPS and P/E ratio.")
else:
    print("There is a weak or no correlation between EPS and P/E ratio.")
```

```python
# Identify companies with high EPS and low P/E ratio
high_eps_low_pe = data[(data['EarningsPerShare'] > data['EarningsPerShare'].quantile(0.75)) & (data['PriceEarningsRatio'] < data['PriceEarningsRatio'].quantile(0.25))]
print("Companies with high EPS and low P/E ratio:")
print(high_eps_low_pe[['Company', 'EarningsPerShare', 'PriceEarningsRatio']])
```

```python
# Investigate potential reasons for high EPS and low P/E ratio
for company in high_eps_low_pe['Company']:
    company_data = data[data['Company'] == company]
    print(f"\nCompany: {company}")
    print(f"Revenue: {company_data['Revenue'].values[0]}")
    print(f"Profit: {company_data['Profit'].values[0]}")
    print(f"Market Capitalization: {company_data['MarketCap'].values[0]}")
    print("Potential reasons could be undervalued stock, high growth potential, or industry-specific factors.")
```

2. **Correlation between EPS and P/E Ratio:**
