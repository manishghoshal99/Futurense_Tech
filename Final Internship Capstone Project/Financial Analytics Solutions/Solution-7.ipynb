```python
import pandas as pd
import plotly.express as px

# Load the dataset
data = pd.read_csv('/Users/manishghoshal/Downloads/Futurense_Internship-main/Final Internship Capstone Project/Financial Analytics/financialanalytics.csv')
```

```python
# Calculate the volatility (standard deviation) of stock price
data['Volatility'] = data.groupby('Company')['PriceEarningsRatio'].transform('std')
```

```python
# Find the company with the highest volatility
highest_volatility_company = data.loc[data['Volatility'] == data['Volatility'].max(), 'Company'].values[0]
highest_volatility_value = data.loc[data['Company'] == highest_volatility_company, 'Volatility'].values[0]
print(f"The company with the highest volatility is {highest_volatility_company} with a volatility of {highest_volatility_value:.2f}")
```

```python
# Analyze the relationship between volatility and market capitalization
fig = px.scatter(data, x='MarketCap', y='Volatility', color='Industry', hover_name='Company',
                 title='Volatility vs. Market Capitalization')
fig.update_layout(xaxis_title='Market Capitalization', yaxis_title='Volatility')
fig.show()
```

```python
# Compare the highest volatility company with others in terms of revenue and profit
highest_volatility_company_data = data.loc[data['Company'] == highest_volatility_company]
print(f"Revenue of {highest_volatility_company}: {highest_volatility_company_data['Revenue'].values[0]}")
print(f"Profit of {highest_volatility_company}: {highest_volatility_company_data['Profit'].values[0]}")
```

```python
# Create a bar chart to compare revenue and profit with other companies
fig = px.bar(data, x='Company', y=['Revenue', 'Profit'], barmode='group', title='Revenue and Profit Comparison')
fig.show()
```

**Infer findings:**

1. **Highest Volatility Analysis:**
   - The company 'B' has the highest volatility with a standard deviation of 0.1308.

2. **Relationship between Volatility and Market Capitalization:**
   - There seems to be a mixed relationship between volatility and market capitalization. Some companies with higher market capitalization have lower volatility, while others show higher volatility.

3. **Comparison with Average Revenue and Profit:**
   - The company with the highest volatility has above-average revenue.
   - The company with the highest volatility has below-average profit.
